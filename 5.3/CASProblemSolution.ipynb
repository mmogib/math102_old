```julia
t = symbols("t", real=true)
f(t) = t==0 ? 1 : sin(t)/t
Si(x) =integrate(f(t),(t,0,x))
nSi(x) = convert(Float64,integrate(f(t),(t,0,x)).n())
Si(1).n()
```

```julia
x = range(0,10,step=0.1)
yy = nSi.(x)
```

```julia
plot(x,yy, label=L"y=Si(x)=\int_0^x \frac{\sin t}{t} dt", legend=:bottomright)
```
