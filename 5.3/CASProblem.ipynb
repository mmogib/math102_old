```julia
using Plots, SymPy, LaTeXStrings
```

```julia
t = symbols("t", real=true)
f(τ) = (τ==0) ? 1 : sin(τ)/τ
Si(x) = integrate(f(t),(t,0,x))
nSi(x) = convert(Float64,Si(x).n())
```

## Part a
Plot

```julia
xx = range(0,10,step=0.1)
yy = nSi.(xx)
```

```julia
plot(xx,yy, label=L"y=Si(x)=\int_0^x \frac{\sin t}{t}dt", legend=:bottomright)
```

## Part b

```julia
solve(f(t),t)
```

The maximum occurs on $x$ where $x=(2n+1)\pi, n=1,2,4.\cdots$ 

## Part c
$Si''(x)=0$

```julia
nddSi(x) = tan(x)-x
solve(nddSi(t),t)
```

```julia
plot(xx,tan.(xx))
plot!(xx,xx)
```
