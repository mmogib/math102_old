Q1) Use the method of cylindrical shells to write (\emph{DO NOT EVALUATE}) an integral for the volume of the solid generated by rotating the region bounded by the curves
\\[
y=-x^2-x\;\; \text{ and } y=0
\\]
about the line $x=1$.

```python
from sympy import plot_implicit, symbols,init_printing, Eq, And, expand, latex
init_printing()
x, y =symbols('x y')
p1 = plot_implicit(Eq(y,-x**2-x),  (x, -2, 2), (y, -1, 1),title="The curve $y=-x^2-x$", show=False)
p2 = plot_implicit(y<-x**2-x,  (x, -1, 0), (y, 0, 1), show=False)
p3 = plot_implicit(Eq(x,1), show=False)
p1.append(p3[0])
p1.append(p2[0])
p1.show()
```

__Q1 `solution`__
$$
r=1+x, \quad h=-x^2-x
$$
So the volume of the solid is found by evaluation the integral
$$
V= \int_{-1}^0 2\pi (1+x)(-x^2-x) dx =\int_{-1}^0 2\pi (- x^{3} - 2 x^{2} - x) dx 
$$

Q2) Find the average value of the function
\\[
f(x) = \sin^{-1} x
\\]
on the interval $[0, 1]$.

__Q2 `solution`__
$$
{\large
\begin{array}{lll}
f_{\text{avg}}&=&\frac{1}{1-0}\int_0^1 \sin^{-1} x dx  \\
              &=& \int_0^1 \sin^{-1} x dx  
\end{array}
}
$$

<div style="position:absolute; right: 10px; top:10px; padding: 5px; margin: 0px; width: 200px; border: 1px solid blue;">
Now, use the substitution  $u=\sin^{-1} x$ and therefore $x= \sin(u)$ and <br />
$$
dx =\cos u du
$$
</div>
So, 
$$
{\large
\begin{array}{lll}
f_{\text{avg}} &=& \int_0^1 \sin^{-1} x dx  \\
                &=& \int_0^{\pi/2} u \cos u du \\
\end{array}
}
$$

now, by parts, we get


<div style="position:absolute; right: 5px; top:10px; padding: 5px; margin: 0px; width: 250px; border: 1px solid blue;">
$$
{\large
\begin{array}{lll}
    y = u, & dz = \cos u du \\
    dy = du, & z = \sin u  \\
\end{array}
}
$$
</div>
$$
{\large
\begin{array}{lll}
\int_0^{\pi/2} u \cos u du  &=& \left. u\sin u \right]_0^{\pi/2} - \int_0^{\pi/2} \sin u du  \\
  &=& \left. u\sin u \right]_0^{\pi/2} + \left. \cos u \right]_0^{\pi/2}   \\
  &=& \frac{\pi}{2} -1  \\
\end{array}
}
$$
Therefore,
$$
f_{\text{avg}} = \frac{\pi}{2} -1 
$$
