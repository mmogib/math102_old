## Question 1
$$\int \frac{2^x}{4^x+1}dx$$

__Solution__
Let $u=2^x$, and so $du = 2^x \ln 2 dx  $
$$\begin{array}{lll} 
\int \frac{2^x}{4^x+1}dx & = & 
    \frac{1}{\ln 2}\int \frac{du}{u^2+1}\\
  & = & 
    \frac{1}{\ln 2}\tan^{-1}u + C \\
  & = &   \frac{1}{\ln 2}\tan^{-1}2^x + C
\end{array}
$$

## Question 2
$$\int \frac{\sec^2x}{\tan^2x\;\;+3\tan x\;\;+2}\;\; dx $$

__Solution__
Let $u=\tan x $, and so $du = \sec^2 x dx $
$$\begin{array}{lll} 
\int \frac{\sec^2x}{\tan^2x\;\;+3\tan x\;\;+2}\;\; dx & = & 
    \int \frac{du}{u^2+3u +1}
\end{array}
$$
By partial fractions
$$ \frac{1}{u^2+3u +1}=\frac{1}{(u+1)(u+2)} = \frac{1}{u+1} - \frac{1}{u+2}$$
So,
$$\begin{array}{lll} 
\int \frac{\sec^2x}{\tan^2x\;\;+3\tan x\;\;+2}\;\; dx & = & 
    \int \frac{du}{u^2+3u +1} \\
    & = & \int \frac{du}{u+1} - \int \frac{du}{u+2} \\
    & = & \ln{|u+1|} - \ln{|u+2|} + C \\
    & = & \ln{\left|\frac{u+1}{u+2}\right|} + C \\
    & = & \ln{\left|\frac{\tan x \; \; + \; \; 1}{\tan x \; \; + \; \; 2}\right|} + C \\
\end{array}
$$

## Question 3
$$\int_{-1}^0 \sqrt{x^2+2x+2} \;\; dx$$

__Solution__
First we complete the square $x^2+2x+2 = x^2+2x+1 +\;\;1=(x+1)^2 +1  $
Let $u=2^x$, and so $du = 2^x \ln 2 dx  $
$$\begin{array}{lll} 
\int_{-1}^0 \sqrt{x^2+2x+2} \;\; dx & = & \int_{-1}^0 \sqrt{(x+1)^2 +1} \;\; dx\\  
    & \overset{u=x+1}{=\mathrel{\mkern-2mu}=} & \int_{0}^1 \sqrt{u^2 +1} \;\; dx\\  
    & \overset{u=\tan{\theta}}{=\mathrel{\mkern-2mu}=} & \int_{0}^{\pi \over 4} \sec^3 \theta \;\; d\theta\\  
    & = & \int_0^{\pi \over 4} \sec^2 \theta \sec \theta d\theta
\end{array}
$$
 
By parts,  
$$
\begin{array}{lll} 
    y = \sec \theta &, & dz = \sec^2 \theta d\theta ;\\
    dy = \tan \theta \sec \theta d\theta &, & z = \tan \theta ; 
\end{array}
$$
So, 
$$
\int_0^{\pi \over 4} \sec^2 \theta \sec \theta d\theta = \left. \sec \theta \tan\theta \right|_0^{\pi \over 4} - \int_0^{\pi \over 4} \sec \theta \tan^2 \theta d\theta =
\left. \sec \theta \tan\theta \right|_0^{\pi \over 4} - \int_0^{\pi \over 4} \sec \theta \left(1+\sec^2\right) \theta d\theta
$$
and 

$$
2 \int_0^{\pi \over 4} \sec^3 \theta  d\theta = \left. \sec \theta \tan\theta \right|_0^{\pi \over 4}-\int_0^{\pi \over 4} \sec \theta =
 \left. \sec \theta \tan\theta \right|_0^{\pi \over 4}- \left. \ln | \sec \theta + \tan \theta \right|_0^{\pi \over 4} = \sqrt{2} - \ln(\sqrt{2} + 1)
$$
therefore, 
$$
 \int_{-1}^0 \sqrt{x^2+2x+2} \;\; dx =\int_0^{\pi \over 4} \sec^3 \theta  d\theta =  \frac{\sqrt{2}}{2} - \frac{\ln(\sqrt{2} + 1)}{2}
$$

## Question 4
$$\int\cos x \; \cos^5(\sin x)\;\;dx$$

__Solution__
Let $u=\sin x$, and so $du = \cos x  dx  $
$$\begin{array}{lll} 
\int\cos x \; \cos^5(\sin x)\;\;dx & = & 
    \int \cos^5(u)\;\;du = \int \cos^4(u)\cos u \;\;du\\
  & = & 
    \int (1-\sin^2 u)^2\cos u \;\;du \\
  & \overset{y=\sin{u}}{=\mathrel{\mkern-2mu}=}  &   \int (1-y^2)^2\;\;dy \\
  & = & \int (1-2y^2+y^4)\;\;dy \\
  & = &  y-{2\over 3} y^3 + {1\over 5} y^5 + C  \\
  & = &  \sin{u}-{2\over 3} \sin^3{u} + {1\over 5} \sin^5{u} + C  \\
  & = &  \sin{\left(\sin x\right)}-{2\over 3} \sin^3{\left(\sin x\right)} + {1\over 5} \sin^5{\left(\sin x\right)} + C  \\
\end{array}
$$
