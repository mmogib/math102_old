```julia
using Plots, Interact, WebIO
gr()
```

```julia
f(x) =1/(1+x^2);
a = 0; b = 1; N = 10
n = 10 # Use n*N+1 points to plot the function smoothly


plot()
@manipulate for N=[10, 30, 50]
    Δx =(b-a)/N;
    x = range(a,b,length=N+1)
    y = f.(x)

    X = range(a,b;length=n*N+1)
    Y = f.(X);
    x_left = (Δx/2).+x[1:end-1] # Left endpoints
    y_left = y[1:end-1]
    left_sum = round(sum(y_left)*Δx,sigdigits=5)

    bar(x_left,y_left,bar_width=Δx,fillalpha=0.2, bar_edges=true,label=nothing);
    p1 = plot!(X,Y, label="Left Riemann Sum, N = $N,\n ∑f(x*)Δx = $left_sum")


    x_right = -(Δx/2).+x[2:end] # Right endpoints
    y_right = y[2:end]
    right_sum = round(sum(y_right)*Δx,sigdigits=5)
    
    bar(x_right,y_right,bar_width=Δx,fillalpha=0.2, bar_edges=true,label=nothing);
    p2 = plot!(X,Y, label="Right Riemann Sum, N = $N, \n ∑f(x*)Δx = $right_sum")

    x_mid = (x[1:end-1].+x[2:end])./2 # Midpoints
    y_mid = f.(x_mid)
    mid_sum = round(sum(y_mid)*Δx, sigdigits=5)
    bar(x_mid,y_mid,bar_width=Δx,fillalpha=0.2, bar_edges=true,label=nothing);
    p3 = plot!(X,Y, label="Midpoint Riemann Sum, N = $N\n ∑f(x*)Δx = $mid_sum")

    plot(p1,p2,p3, layout=3)
end
```
