```julia
using SymPy, Plots, LaTeXStrings
gr()
```

```julia
x = symbols("x", real=true)
```

```julia
f(x)=x/sqrt(x^2+1)
f(x)
```

```julia
g(x)=x^4-x
g(x)
```

```julia
poi = convert(Array{Float64,2},solve(g(x)-f(x),x).n())
```

```julia
p1=plot(f(x),xlims=(-1,2),ylims=(-1,2), label=L"f(x)=\frac{x}{\sqrt{x^2+1}}")
p2 =plot!(p1,g(x),xlims=(-1,2),ylims=(-1,2),label=L"g(x)=x^4-x")
plot!(p2,poi,f.(poi),seriestype = :scatter, 
    label="intersection points",ys=poi, gridalpha=0.5, 
    minorgrid=true, minorgridalpha=0.5, framestyle = :zerolines )
```

```julia
integrate(f(x)-g(x),(x,poi[1],poi[2]))
```
